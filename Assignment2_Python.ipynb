```python
#matrix multiplication investigation

process = psutil.Process(os.getpid())
baseRam = process.memory_info().rss

resTime = np.zeros(10)
resSpace = np.zeros(10)

for i in range(1,11):
    gc.collect()
    start = time.time()
    cl.matrix_multiplication(i*1000)
    end = time.time()
    ram = process.memory_info().rss    
    resTime[i-1] = end - start
    resSpace[i-1] = ram - baseRam
```

```python
size=np.linspace(1000,10000,10)
```

```python
data=np.asarray([size,resTime,resSpace]).T
columns=['Size','Time','Space']
```

```python
lf=pd.DataFrame(data=data,columns=columns)
```

```python
lf.plot(x='Size',y='Time',title='Time against Size')#parabolic increase in time is seen below
```

```python
lf.plot(x='Size',y='Space',title='Space against Size')
#no clear pattern in space used but increase in space is not exponentially increasing with size
#More variables seem to affect the space used that cannot be controlled
```

```python
#integer sorting investigation
process = psutil.Process(os.getpid())
baseRam = process.memory_info().rss

resTime = np.zeros(10)
resSpace = np.zeros(10)

for i in range(1,11):
    gc.collect()
    start = time.time()
    cl.int_sorting(i*1000)
    end = time.time()
    ram = process.memory_info().rss    
    resTime[i-1] = end - start
    resSpace[i-1] = ram - baseRam
```

```python
size=np.linspace(1000,10000,10)
data=np.asarray([size,resTime,resSpace]).T
columns=['Size','Time','Space']
lf=pd.DataFrame(data=data,columns=columns)
```

```python

#substring loop-based implementation

process = psutil.Process(os.getpid())
baseRam = process.memory_info().rss

resTime = np.zeros(10)
resSpace = np.zeros(10)

string='abcdefghijklmnopqrstuvwxyz'

for i in range(1,11):
    gc.collect()
    start = time.time()
    txt="".join(string[np.random.randint(0,26, (i*1000))[j]] for j in range(i*1000))
    lookup="".join(string[np.random.randint(0,26, (10))[j]] for j in range(10))
    cl.subset_string(txt,lookup)
    end = time.time()
    ram = process.memory_info().rss    
    resTime[i-1] = end - start
    resSpace[i-1] = ram - baseRam
```

```python
#implementation with python function


process = psutil.Process(os.getpid())
baseRam = process.memory_info().rss

resTime_python = np.zeros(10)
resSpace_python = np.zeros(10)

for i in range(1,11):
    gc.collect()
    start = time.time()
    txt="".join(string[np.random.randint(0,26, (i*1000))[j]] for j in range(i*1000))
    lookup="".join(string[np.random.randint(0,26, (10))[j]] for j in range(10))
    x=txt.find(lookup)
    end = time.time()
    ram = process.memory_info().rss    
    resTime_python[i-1] = end - start
    resSpace_python[i-1] = ram - baseRam
```

```python
size=np.linspace(1000,10000,10)
data=np.asarray([size,resTime,resTime_python,resSpace,resSpace_python]).T
columns=['Size','Time with loop','Time with Python','Space with loop','Space with python']
lf=pd.DataFrame(data=data,columns=columns)
```

```python
lf.plot(x='Size',y=['Time with loop','Time with Python'])
#parabolic increase in time against size is seen below with almost identical performance for both
```

```python
lf.plot(x='Size',y=['Space with loop','Space with python'])
#performance can vary when run at different times between 2 methods
#In the case shown below space with python remains constant at the most part after size=3000, similar pattern for space with loop 
#python performs much better in space used
#More variables seem to affect the space used that cannot be controlled
```
